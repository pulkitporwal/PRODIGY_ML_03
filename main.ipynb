```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import cv2
import os
import pickle
import random
from sklearn.svm import SVC
```

```python
dir = 'C:\\Users\\Lenovo\\OneDrive\\Desktop\\Pulkit_Porwal\\data\\train'
print(dir)
```

```python
categories = ['Cat', 'Dog']
data = []
```

```python
try:
    with open('data.pickle', 'rb') as pick_in:
        data = pickle.load(pick_in)
    print("Loaded data from data.pickle.")

except FileNotFoundError:
    print("No existing data.pickle file found. Generating new data.")
    for category in categories:
        path = os.path.join(dir, category)
        label = categories.index(category)

        print(path)
        print(label)

        for img in os.listdir(path):
            imgpath = os.path.join(path, img)
            image = cv2.imread(imgpath, 0)
            image = cv2.resize(image, (50, 50))  
            image_data = np.array(image).flatten()

            data.append([image_data, label])

    with open('data.pickle', 'wb') as pick_out:
        pickle.dump(data, pick_out)
    print("Data saved to data.pickle.")
```

```python
data
```

```python
random.shuffle(data)
```

```python
features = []
labels = []
for feature, label in data:
    features.append(feature)
    labels.append(label)

max_feature_length = len(features[0])  
x_train = np.zeros((len(features), max_feature_length), dtype=np.uint8)
y_train = np.array(labels)

```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=20)
```

```python
model = SVC(kernel="poly", gamma='auto')
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
accuracy = model.score(predictions,y_test)
```
